```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def Binmoal(n, sec):
  lisy = []
  lisx = []
  for i in range(0, n+1):
    lisx.append(i)
    p = pow(sec, i)
    x = 1-p
    q = pow(x, n-i)
    c = math.comb(n, i)
    lisy.append(c*q*p)
  return lisx, lisy


```

```python
x , y = Binmoal(10 , 0.3)
for i in x:
  print(i , ': ', y[i])
```

```python
plt.plot(x,y)
plt.show()
```
