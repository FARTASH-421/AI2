```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
def binomial_pro(n, M, N):
  lisIndex = []
  listBin = []
  for k in range(0, n+1):
    comb = math.comb(n, k)
    y = M/N
    p = pow(y, k)
    q = pow((1-y), n-k)
    result = comb * p * q

    lisIndex.append(k)
    listBin.append(result)

  return lisIndex, listBin
```

```python
def Hypergeometric_dis(n, M, N):
  list_x = []
  list_result = []
  for x in range(0, n+1):
    comb1 = math.comb(M, x)
    comb2 = math.comb(N-M, n-x)
    comb3 = math.comb(N, n)
    result = (comb1 * comb2) / comb3
    list_x.append(x)
    list_result.append(result)

  return list_x, list_result
```

```python
k, r = binomial_pro(10, 20, 100)
k1, r1 = Hypergeometric_dis(10, 20, 100)

# plt.plot(k, r, k1, r1)
# plt.grid()
# plt.show()
```

```python
# plot binomail
plt.subplot(2, 1, 1)
plt.plot(k, r, color='r', linewidth = '2')
plt.title("Binomial Dis")
plt.xlabel('x')
plt.ylabel("f(x)")
plt.grid()


#plot Hypergeometric
plt.subplot(2, 1, 2)
plt.plot(k1,r1, color = 'g', linewidth = '2')
plt.title("Hypergeometric Dis")
plt.xlabel('x')
plt.ylabel("f(x)")
plt.grid()


plt.show()
```
