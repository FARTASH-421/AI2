```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img.jpeg');
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap="gray")
plt.show()
```

```python
def Gama(image, alpha, c):
  new_img = np.zeros(image.shape)
  x , y = image.shape
  for i in range(0, x):
    for j in range(0, y):
      d = image[i][j]
      new_img[i][j] = c* pow(d, alpha)


  return new_img

```

```python
new = Gama(gray,1, 1)
plt.imshow(new, cmap='gray')
plt.show()
```

```python
new = Gama(gray,2, 2)
plt.imshow(new, cmap='gray')
plt.show()
```

```python
new = Gama(gray,2, 1)
plt.imshow(new, cmap='gray')
plt.show()
```

```python
new = Gama(gray,4, 1)
plt.imshow(new, cmap='gray')
plt.show()
```
